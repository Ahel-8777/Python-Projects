```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('data.csv.txt')
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
#Finding the missing values in the columns
df.isnull().sum()
```

```python
#Filling missing values
df['Calories'].fillna(df['Calories'].median(), inplace=True)
```

```python
#Ploting the values using seaborn data visualization library.
sns.pairplot(df, height=2.5);
```

```python
import warnings
warnings.filterwarnings('ignore')

plt.figure(figsize=(4, 3.5))
sns.distplot(df['Pulse']);
```

```python
plt.figure(figsize=(4, 3.5))
sns.distplot(df['Maxpulse']);
```

```python
plt.figure(figsize=(5, 3.5))
plt.scatter(df['Maxpulse'], df['Duration'], color='red', alpha=0.5)
plt.title('Maxpulse by duration', fontsize=13)
plt.xlabel('Maxpulse', fontsize=10)
plt.ylabel('Duration', fontsize=10)
plt.axhline(y=df['Duration'].mean(), linestyle='--', linewidth=1.0, color='black');
```

```python
#Co-Relation Matrix
plt.figure(figsize=(4, 4))
cor_mat=df.corr()
cmap=sns.light_palette('seagreen', as_cmap=True)
sns.heatmap(cor_mat, cmap=cmap, cbar=False, annot=True, linewidth=0.5, linecolor='salmon');
```
